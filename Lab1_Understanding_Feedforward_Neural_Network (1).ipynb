#### Part 1 data preparation

In the cell below, generate training data for your model.
<br> Generate a decent amount of training data in the interval of x.
<br>
<br>All data need to be separated by the same distance.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the interval and separation distance
x_min, x_max = -100, 100
num_points = 500  # You can adjust this based on your desired amount of training data
delta_x = (x_max - x_min) / (num_points - 1)

# Generate x values
x_values = np.linspace(x_min, x_max, num_points)

# Calculate corresponding y values using the given function
y_values = x_values * np.sin(x_values**2 / 300)

# Visualize the data (optional)
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r"$y = x \sin\left(\frac{x^2}{300}\right)$")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Training Data for y = x sin(x^2/300)")
plt.grid(True)
plt.legend()
plt.show()

# Now you can use (x_values, y_values) as your training data for each of your FNNs.



```

#### Part 2 build models

In the cells below, build your models and train it with data from part1.
<br>You need to split your training data to two parts. With 40% used in training, and 60% used in test.

<br>You may do this with sklearn's train_test_split() or keras' validation_split together with train_test_split().

```python
# Import necessary libraries
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Split the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.6, random_state=42)

# Model 1
model1 = Sequential([
    Dense(64, input_shape=(1,), activation='relu'),
    Dense(32, activation='relu'),
    Dense(1)
])

# Compile and train Model 1
model1.compile(optimizer='adam', loss='mean_squared_error')
model1.fit(x_train, y_train, epochs=50, batch_size=32, validation_data=(x_test, y_test))
```

```python
# model 2
# Import necessary libraries
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Split the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.6, random_state=42)


model2 = Sequential([
    Dense(128, input_shape=(1,), activation='sigmoid'),
    Dense(64, activation='sigmoid'),
    Dense(1)
])

model2.compile(optimizer='adam', loss='mean_squared_error')
model2.fit(x_train, y_train, epochs=50, batch_size=32, validation_data=(x_test, y_test))

```

```python
# model 3
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Split the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.6, random_state=42)

# Model 3
model3 = Sequential([
    Dense(32, input_shape=(1,), activation='tanh'),
    Dense(16, activation='tanh'),
    Dense(1)
])

# Compile and train Model 3
model3.compile(optimizer='adam', loss='mean_squared_error')
model3.fit(x_train, y_train, epochs=50, batch_size=32, validation_data=(x_test, y_test))

```

#### Part 3 model evaluation

In the cells below, eval your model with training data, test data ( with model.evaluate( ) ), and plot your predict result on the same plot with the plot of goal function.

```python
# Evaluate and plot results for Model 1
train_loss_model1 = model1.evaluate(x_train, y_train, verbose=0)
test_loss_model1 = model1.evaluate(x_test, y_test, verbose=0)

print("Model 1 - Training Loss:", train_loss_model1)
print("Model 1 - Test Loss:", test_loss_model1)

# Plot predicted results for Model 1
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r"$y = x \sin\left(\frac{x^2}{300}\right)$", color='blue')
plt.scatter(x_test, model1.predict(x_test), label="Model 1 Predictions", color='red', alpha=0.5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Model 1 Evaluation and Predictions")
plt.grid(True)
plt.legend()
plt.show()

# Evaluate and plot results for Model 2
train_loss_model2 = model2.evaluate(x_train, y_train, verbose=0)
test_loss_model2 = model2.evaluate(x_test, y_test, verbose=0)

print("Model 2 - Training Loss:", train_loss_model2)
print("Model 2 - Test Loss:", test_loss_model2)

# Plot predicted results for Model 2
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r"$y = x \sin\left(\frac{x^2}{300}\right)$", color='blue')
plt.scatter(x_test, model2.predict(x_test), label="Model 2 Predictions", color='green', alpha=0.5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Model 2 Evaluation and Predictions")
plt.grid(True)
plt.legend()
plt.show()

# Evaluate and plot results for Model 3
train_loss_model3 = model3.evaluate(x_train, y_train, verbose=0)
test_loss_model3 = model3.evaluate(x_test, y_test, verbose=0)

print("Model 3 - Training Loss:", train_loss_model3)
print("Model 3 - Test Loss:", test_loss_model3)

# Plot predicted results for Model 3
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r"$y = x \sin\left(\frac{x^2}{300}\right)$", color='blue')
plt.scatter(x_test, model3.predict(x_test), label="Model 3 Predictions", color='orange', alpha=0.5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Model 3 Evaluation and Predictions")
plt.grid(True)
plt.legend()
plt.show()



```

#### Part 4 get model output and feedforward by yourself

Recall how a Feedforward Neural Network gets its output. Now choose your model with highest accuracy, and call get_weights( ) to get its weights and bias.
<br> Hint: bias and weights have different dimensions in most cases. Try to guess which index represent bias before you look it up on the internet.

<br> Afterwards, choose 5 data from your training dataset, do all the necessary calculations (with program of course), and get the output of your model. Compare it to the model.predict( ) result.
<br> They should be identical for the first several digitals, if not for all digits.

```python
# Get weights and biases from Model 1
weights_model1 = model1.get_weights()

# Choose 5 data points from the training dataset
x_sample = x_train[:5]

# Initialize the input
input_data = x_sample

# Feedforward calculation
for i in range(len(weights_model1) // 2):
    # Get weights and biases for the current layer
    weights = weights_model1[i*2]
    biases = weights_model1[i*2 + 1]

    # Calculate the output of the current layer
    output_data = np.dot(input_data, weights) + biases

    # Apply ReLU activation function for hidden layers
    if i < len(weights_model1) // 2 - 1:
        output_data = np.maximum(output_data, 0)

    # The output of the current layer is the input to the next layer
    input_data = output_data

# Print the final output
print("The output of the model for the 5 data points is:", output_data)



```
